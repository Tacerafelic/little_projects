# Mandelbrot $z^2$

```python
import numpy as np
import matplotlib.pyplot as plt


def mandelbrot(c, max_iteration):
    z = 0
    for n in range(max_iteration):
        if abs(z) > 2:
            return n
        z = z ** 2 + c
    return max_iteration


def mandelbrot_set(x_min, x_max, y_min, y_max, width, height, max_iteration):
    x = np.linspace(x_min, x_max, width)
    y = np.linspace(y_min, y_max, width)
    mandeln = np.zeros((height, width))

    for i in range(height):
        for j in range(width):
            c = complex(x[j], y[i])
            mandeln[i, j] = mandelbrot(c, max_iteration)
    return mandeln


x_min, x_max, y_min, y_max = -2.0, 1.0, -1.5, 1.5
width, height = 1000, 1000
max_iteration = 100

mandelbrot_image = mandelbrot_set(x_min, x_max, y_min, y_max, width, height, max_iteration)

fig = plt.figure()
plt.imshow(mandelbrot_image, extent=[x_min, x_max, y_min, y_max], cmap="twilight_shifted")
plt.colorbar()
plt.title("Mandelbrot Set")
plt.xlabel("Re(c)")
plt.ylabel("Im(c)")
plt.show()
# fig.savefig('Mandelbrot_400.pdf')
```
